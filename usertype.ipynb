```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('201801_hubway_tripdata.csv')


```

```python
df.tripduration = df.tripduration/60
```

```python
subs=0
tourist = 0
total_ride_subs=0
total_ride_tourists=0

for i in range(0,len(df)):
    if(df.usertype[i] == 'Subscriber'):
        subs=subs+1
        total_ride_subs = total_ride_subs + df.tripduration[i]
        
    else:
        tourist = tourist+1
        total_ride_tourists = total_ride_tourists + df.tripduration[i]
#print(total_ride_subs,total_ride_tourists)
#print(subs,tourist)
average_ride_subscriber = (total_ride_subs) / subs
average_ride_tourists = total_ride_tourists / tourist

#print (average_ride_subscriber, average_ride_tourists)

import matplotlib.pyplot as plt

#print("Distribution of Types of Users")
print(" ")
# Plotting Distribution OF USERS
labels = 'Subscribers', 'One-Timers'
sizes = [subs, tourist]
colors = ['lightskyblue', 'red']

plt.pie(sizes, labels=labels, colors=colors,
        autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.show()




```
