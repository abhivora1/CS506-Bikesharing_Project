```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("income_neighborhood.csv")
df
```

```python
df = df.dropna()

df_nos_of_stations = df.sort_values(by=['number_of_stations'], ascending=False)
df_nos_of_stations.to_csv("sorted_based_on_stations.csv")

df_based_on_income = df.sort_values(by=['Average_Income'], ascending=False)
df_based_on_income.to_csv("sorted_based_on_income.csv")
print(df_based_on_income)

```
